### Open Bottle neck features

```python
import numpy as np
from keras.callbacks import ModelCheckpoint   
from keras.models import Sequential
from keras.layers import Dense, Dropout, Conv2D, GlobalAveragePooling2D,BatchNormalization,Activation
from sklearn.datasets import load_files      
from keras.utils import np_utils
from keras.applications import VGG16
import scipy.misc
from keras.applications.vgg16 import preprocess_input
from glob import glob
import matplotlib.pyplot as plt
import cv2
```

```python
bottleneck_features = np.load('bottleneck_features/DogVGG16Data.npz')
```

### Making labels

```python
train_VGG16 = bottleneck_features['train']
valid_VGG16 = bottleneck_features['valid']
test_VGG16 = bottleneck_features['test']
```

```python
valdata=load_files('/home/shivam/Dog Breed Identification/dogImages/valid')
traindata=load_files('/home/shivam/Dog Breed Identification/dogImages/train')
testdata=load_files('/home/shivam/Dog Breed Identification/dogImages/test')
```

```python
train_targets = np_utils.to_categorical(np.array(traindata['target']), 133)
test_targets = np_utils.to_categorical(np.array(testdata['target']), 133)
valid_targets = np_utils.to_categorical(np.array(valdata['target']), 133)
```

```python
print "Data has "+str(train_VGG16.shape[0]+test_VGG16.shape[0]+valid_VGG16.shape[0])+" dog images with 133 categories"
```

```python
print "Train has "+ str(train_VGG16.shape[0])+" elements"
print "Test has "+str(test_VGG16.shape[0])+" elements"
print "Valid has "+str(valid_VGG16.shape[0])+" elements"
```

```python
print "Shape of bottleneck features is "+str(train_VGG16.shape[1:])
```

```python
VGG16_model = Sequential()
VGG16_model.add(GlobalAveragePooling2D(input_shape=train_VGG16.shape[1:]))
VGG16_model.add(Dropout(0.8))
VGG16_model.add(Dense(150))
VGG16_model.add(BatchNormalization())
VGG16_model.add(Activation('relu'))
VGG16_model.add(Dense(133, activation='softmax'))

VGG16_model.summary()
```

```python
VGG16_model.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
```

```python
checkpointer = ModelCheckpoint(filepath='saved_models/weights.best.VGG16.hdf5', 
                               verbose=1, save_best_only=True)
VGG16_model.fit(train_VGG16, train_targets, 
          validation_data=(valid_VGG16, valid_targets),
          epochs=50, batch_size=20, callbacks=[checkpointer], verbose=1)
```

### Test prediction

```python
VGG16_model.load_weights('saved_models/weights.best.VGG16.hdf5')

# evaluate test accuracy
score = VGG16_model.evaluate(test_VGG16, test_targets, verbose=0)
accuracy = 100*score[1]

# print test accuracy
print('Test accuracy: %.4f%%' % accuracy)
```

###  Dog breed prediction

```python
input_path='/home/shivam/Dog Breed Identification/dogImages/train/002.Afghan_hound/Afghan_hound_00082.jpg'
```

```python
dog_names = [item[20:-1] for item in sorted(glob("dogImages/train/*/"))]#list of dog names
```

```python
def calculate_bottleneck_features(path):
    model=VGG16(include_top=False,weights='imagenet',input_shape=(224,224,3))
    image=scipy.misc.imread(path)
    image=scipy.misc.imresize(image,(224,224,3))
    image=np.array([image]).astype('float32')
#     print image.shape
    model.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
    final_bottleneck=model.predict(preprocess_input(image))
    return final_bottleneck
    
    
```

```python
a=calculate_bottleneck_features(input_path)
```

### Save your VGG model in json

```python
with open('model_architecture_vgg_self_made.json', 'w') as f:
    f.write(VGG16_model.to_json())
```

```python
dog_breed=dog_names[np.argmax(VGG16_model.predict(a),axis=1)[0]]
```
